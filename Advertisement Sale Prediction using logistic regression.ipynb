```python
# Predicting whether the customer is going to buy the product or not
# so it comes under supervised learning.
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Book1.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
# segregate dataset into X(input) and Y(output)
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
#splitting dataset into train and test
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
# future Scaling
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train=sc.fit_transform(x_train)
x_test= sc.transform(x_test)
print(x_train)
```

```python
#training
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
# prediction of data
```

```python
y_pred=model.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
# the first column is predicted value and second one is tested value
```

```python
# Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm= confusion_matrix(y_test,y_pred)
print('Confusion Matrix: ')
print(cm)
print("Accuracy of the model: {0}%".format(accuracy_score(y_test,y_pred)*100))
```

```python
# prediction
```

```python
age = int(input("Enter New Customers Age:"))
sal=int(input("enter new customers salary"))
newcust=[[age,sal]]
result=model.predict(sc.transform(newcust))
print(result)
if result == 1:
    print('customer will buy')
else:
    print('customer will not buy')
```
